```python
import os

NOTEBOOK_DIR = os.path.abspath('')
TASK2_DIR    = os.path.abspath(os.path.join(NOTEBOOK_DIR, '..'))
DATA_DIR     = os.path.join(TASK2_DIR, 'data')

print(f"Notebook dir : {NOTEBOOK_DIR}")
print(f"Task2 dir    : {TASK2_DIR}")
print(f"Data dir     : {DATA_DIR}")
```

```python
RAW_IMG_DIR = os.path.join(DATA_DIR, 'raw-img')

for cls in sorted(os.listdir(RAW_IMG_DIR)):
    n = len(os.listdir(os.path.join(RAW_IMG_DIR, cls)))
    print(f"  {cls:<15} {n} images")
```

```python
mapping = {
    "cane":       "dog",
    "gatto":      "cat",
    "mucca":      "cow",
    "elefante":   "elephant",
    "farfalla":   "butterfly",
    "gallina":    "chicken",
    "cavallo":    "horse",
    "pecora":     "sheep",
    "ragno":      "spider",
    "scoiattolo": "squirrel",
}

for italian, english in mapping.items():
    src = os.path.join(RAW_IMG_DIR, italian)
    dst = os.path.join(RAW_IMG_DIR, english)
    if os.path.exists(src):
        os.rename(src, dst)
        print(f"  {italian} -> {english}")
    elif os.path.exists(dst):
        print(f"  {english} already exists, skipping")
    else:
        print(f"  not found: {src}")

for cls in sorted(os.listdir(RAW_IMG_DIR)):
    print(f"  {cls}")
```
